### ðŸŽ² <span style="color:#4682b4;">Introduction</span>

The **Binomial Probability Distribution** describes the likelihood of a specific number of successes in a fixed number of independent experiments, each having the same probability of success. It's used to model scenarios where there are two possible outcomes (often termed as "success" and "failure") in each trial. Examples include flipping a coin, where heads might be considered a success, or determining the number of defective items in a batch of products.

#### **Key Characteristics**:

- **Fixed Number of Trials**: The experiment is repeated a set number of times.
- **Binary Outcomes**: Each trial has only two possible outcomes.
- **Constant Probability**: The probability of success is the same in each trial.
- **Independent Trials**: The outcome of one trial does not affect the others.

This distribution helps in predicting the probability of achieving a certain number of successes over the given number of trials, making it a fundamental tool in statistics and probability theory.

### ðŸŽ² <span style="color:#4682b4;">Understanding Binomial Probability Distribution</span>

**Binomial Probability Distribution** describes the probability of achieving a specific number of successes in a fixed number of independent Bernoulli trials, each with the same probability of success.

#### **Mathematical Definition**:

The binomial probability distribution is defined by the following parameters:
- **Number of trials (n)**: The total number of independent trials.
- **Probability of success (p)**: The probability of success in a single trial.
- **Number of successes (k)**: The specific number of successes for which the probability is calculated.

#### **Binomial Probability Formula**:

The probability of obtaining exactly \( k \) successes in \( n \) trials is given by:

\[ P(X = k) = \binom{n}{k} p^k (1-p)^{n-k} \]

Where:
- \( \binom{n}{k} \) is the binomial coefficient, calculated as:

\[ \binom{n}{k} = \frac{n!}{k!(n-k)!} \]

- \( p^k \) represents the probability of having \( k \) successes.
- \( (1-p)^{n-k} \) represents the probability of having \( n-k \) failures.

#### **Expected Value and Variance**:

The expected value \( E(X) \) and variance \( \text{Var}(X) \) of a binomial distribution are given by:

\[ E(X) = np \]

\[ \text{Var}(X) = np(1-p) \]

These formulas provide the mean and the variability of the distribution, respectively.

#### **Properties of Binomial Distribution**:

- **Discrete Distribution**: The binomial distribution is a discrete probability distribution.
- **Symmetry**: The distribution is symmetric if \( p = 0.5 \); otherwise, it is skewed.
- **Special Case**: When \( n = 1 \), the binomial distribution becomes the Bernoulli distribution.

Understanding these formulas and properties is crucial for analyzing scenarios where binary outcomes are repeated multiple times under identical conditions.

### ðŸŽ² <span style="color:#4682b4;">Practical Examples</span>

The binomial probability distribution can be applied to various real-world scenarios where there are fixed numbers of trials, two possible outcomes for each trial, and constant probabilities of success.

#### **Examples**:

1. **Quality Control in Manufacturing**
    - **Scenario**: A factory produces a large number of light bulbs, and each bulb has a 5% chance of being defective. The probability of finding exactly 3 defective bulbs in a batch of 50 can be modeled using the binomial distribution.

2. **Clinical Trials**
    - **Scenario**: In a clinical trial, a new drug is tested on 100 patients, and each patient has a 70% chance of responding positively to the treatment. The distribution can be used to calculate the probability of exactly 75 patients responding positively.

3. **Customer Support**
    - **Scenario**: A customer support team handles 20 calls per hour, and each call has a 40% chance of being resolved on the first attempt. The binomial distribution can help determine the probability of resolving exactly 8 calls in an hour.

4. **Survey Analysis**
    - **Scenario**: In a survey, 60% of respondents are expected to prefer a new product. Out of 30 respondents, the probability that exactly 18 will prefer the new product can be found using the binomial distribution.

5. **Election Polling**
    - **Scenario**: In an election poll, a candidate has a 55% chance of receiving a vote from any individual voter. The probability of the candidate receiving exactly 600 votes out of 1000 can be calculated using the binomial distribution.

These examples illustrate how the binomial probability distribution is used to model and analyze situations with discrete outcomes and fixed probabilities of success.

### ðŸŽ² <span style="color:#4682b4;">Problem 1</span>

**Problem**:

A factory produces light bulbs, and each bulb has a 4% chance of being defective. If a quality control inspector randomly selects 20 light bulbs, what is the probability that exactly 2 of them are defective?

#### **Solution**:

To solve this problem, we use the binomial probability formula:

$$ P(X = k) = \binom{n}{k} p^k (1-p)^{n-k} $$

Where:
- \( n = 20 \) (number of trials)
- \( p = 0.04 \) (probability of success, i.e., defective bulb)
- \( k = 2 \) (number of successes, i.e., defective bulbs)

First, calculate the binomial coefficient:

$$ \binom{20}{2} = \frac{20!}{2!(20-2)!} = \frac{20 \times 19}{2 \times 1} = 190 $$

Next, calculate the probability:

$$ P(X = 2) = 190 \times (0.04)^2 \times (0.96)^{18} $$

Breaking it down:

$$ (0.04)^2 = 0.0016 $$

$$ (0.96)^{18} \approx 0.4420 $$

Now, multiply these values together:

$$ P(X = 2) = 190 \times 0.0016 \times 0.4420 \approx 0.1347 $$

So, the probability that exactly 2 out of 20 light bulbs are defective is approximately 0.1347 or 13.47%.
