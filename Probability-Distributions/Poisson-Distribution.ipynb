    <h2 style="color: purple;">Poisson Distribution</h2>
    <p style="font-size: 16px; color: black;">
        The Poisson distribution is a discrete probability distribution that describes the number of events occurring in a fixed interval of time or space, given that these events occur with a known constant mean rate and are independent of the time since the last event. It is commonly used to model the number of arrivals in a fixed period of time, such as the number of customers arriving at a service point in a given hour, the number of phone calls received by a call center in a minute, or the number of particles emitted by a radioactive source in a given time interval.
    </p>
    <h3 style="color: purple;">Probability Mass Function (PMF)</h3>
    <p style="font-size: 16px; color: black;">
        The probability mass function (PMF) of the Poisson distribution gives the probability of observing exactly $ k $ events in a fixed interval of time or space, given that the average rate of occurrence of events is $ \lambda $ per interval.
    </p>
    <p style="font-size: 16px; color: black;">
        The PMF of the Poisson distribution is given by the formula:
    </p>
    <p style="font-size: 16px; color: black; margin-left: 40px;">
        $$ P(X = k) = \frac{{e^{-\lambda} \lambda^k}}{{k!}} $$
    </p>
    <h3 style="color: purple;">Cumulative Distribution Function (CDF)</h3>
    <p style="font-size: 16px; color: black;">
        The cumulative distribution function (CDF) of the Poisson distribution gives the probability that the number of events is less than or equal to a certain value $ k $.
    </p>
    <p style="font-size: 16px; color: black;">
        The CDF of the Poisson distribution can be calculated by summing the probabilities of all possible outcomes up to $ k $:
    </p>
    <p style="font-size: 16px; color: black; margin-left: 40px;">
        $$ F(k; \lambda) = \sum_{i=0}^{k} \frac{{e^{-\lambda} \lambda^i}}{{i!}} $$
    </p>
    <h3 style="color: purple;">Mean and Variance</h3>
    <p style="font-size: 16px; color: black;">
        The mean ($ \mu $) and variance ($ \sigma^2 $) of the Poisson distribution are both equal to the average rate of occurrence of events, denoted by $ \lambda $.
    </p>
    <p style="font-size: 16px; color: black;">
        These parameters represent the expected number of events in the given interval and the spread or variability of the distribution, respectively.
    </p>
    <h3 style="color: purple;">Use in Statistics and Data Science</h3>
    <p style="font-size: 16px; color: black;">
        The Poisson distribution is widely used in various fields within statistics and data science:
    </p>
    <ul style="font-size: 16px; color: black;">
        <li><strong>Queuing Theory:</strong> It helps in modeling the number of arrivals at a service point, such as customers at a bank or calls at a call center.</li>
        <li><strong>Risk Management:</strong> Insurance companies use the Poisson distribution to model the number of claims received in a certain period.</li>
        <li><strong>Biostatistics:</strong> It is used to model the occurrence of rare diseases or the number of mutations in a given length of DNA.</li>
        <li><strong>Operations Research:</strong> It helps in analyzing systems and processes where events occur randomly but with a known average rate, such as machine failures.</li>
        <li><strong>Event Detection:</strong> In social media analytics, it can model the occurrence of specific events (like tweets or posts) over time.</li>
        <li><strong>Quality Control:</strong> It is used in manufacturing to model the number of defects or failures in a batch of products.</li>
    </ul>
    <p style="font-size: 16px; color: black;">
        The Poisson distribution's assumption of independence and a constant average rate makes it a versatile tool for modeling count data in various domains.
    </p>

    <h2 style="color: purple;">Example Problems</h2>
    <ul style="font-size: 16px; color: black;">
        <li>
            <strong>Problem 1:</strong> Suppose the average number of emails received per hour is 5. What is the probability of receiving exactly 3 emails in an hour?
            <br><strong>Solution:</strong>
            Here, $ \lambda = 5 $ and $ k = 3 $.
            <br>Using the PMF of the Poisson distribution:
            <br>$$ P(X = 3) = \frac{e^{-5} \cdot 5^3}{3!} = \frac{e^{-5} \cdot 125}{6} \approx 0.1404 $$
            <br>So, the probability of receiving exactly 3 emails in an hour is approximately 0.1404.
        </li>
        <li>
            <strong>Problem 2:</strong> A bookstore sells an average of 2 rare books per week. What is the probability that the bookstore sells at most 1 rare book in a week?
            <br><strong>Solution:</strong>
            Here, $ \lambda = 2 $ and we need to find $ P(X \leq 1) $.
            <br>Using the CDF of the Poisson distribution:
            <br>$$ P(X \leq 1) = P(X = 0) + P(X = 1) $$
            <br>$$ P(X = 0) = \frac{e^{-2} \cdot 2^0}{0!} = e^{-2} $$
            <br>$$ P(X = 1) = \frac{e^{-2} \cdot 2^1}{1!} = 2e^{-2} $$
            <br>$$ P(X \leq 1) = e^{-2} + 2e^{-2} = 3e^{-2} \approx 0.406 $$
            <br>So, the probability that the bookstore sells at most 1 rare book in a week is approximately 0.406.
        </li>
        <li>
            <strong>Problem 3:</strong> A call center receives an average of 10 calls per hour. What is the probability that they receive more than 12 calls in an hour?
            <br><strong>Solution:</strong>
            Here, $ \lambda = 10 $ and we need to find $ P(X > 12) $.
            <br>Using the CDF of the Poisson distribution, we first find $ P(X \leq 12) $:
            <br>$$ P(X \leq 12) = \sum_{i=0}^{12} \frac{e^{-10} \cdot 10^i}{i!} $$
            <br>Using a Poisson CDF table or computational tools, we find:
            <br>$$ P(X \leq 12) \approx 0.7919 $$
            <br>Therefore:
            <br>$$ P(X > 12) = 1 - P(X \leq 12) = 1 - 0.7919 = 0.2081 $$
            <br>So, the probability of receiving more than 12 calls in an hour is approximately 0.2081.
        </li>
    </ul>

    <h2 style="color: purple;">Poisson Probability Distribution Curve </h2>

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the Poisson distribution
lambda_x = 5  # Average rate for x-axis
lambda_y = 7  # Average rate for y-axis

# Number of data points
num_points = 1000

# Generate Poisson-distributed data points
x_data = np.random.poisson(lambda_x, num_points)
y_data = np.random.poisson(lambda_y, num_points)

# Plot the data points
plt.figure(figsize=(10, 6))
plt.scatter(x_data, y_data, alpha=0.5, edgecolors='w', linewidth=0.5)
plt.title('2D Poisson Distribution')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)

# Show the plot
plt.show()
```
